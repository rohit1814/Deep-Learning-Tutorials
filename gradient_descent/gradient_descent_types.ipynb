```python
import numpy as np
import pandas as pd
import time
```

```python
df = pd.read_csv('Social_Network_Ads.csv')
```

```python
df.head()
```

```python
df = df[['Age','EstimatedSalary','Purchased']]
```

```python
df.head()
```

```python
X = df.iloc[:,0:2]
y = df.iloc[:,-1]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

```python
X_scaled.shape
```

```python
#from sklearn.model_selection import train_test_split
#X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
#X_train.shape
```

```python
import tensorflow as tf
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
```

```python
model = Sequential()

model.add(Dense(10,activation='relu',input_dim=2))
model.add(Dense(10,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

#### From the above two results you can see
- Convergence is fater in Stochastic  --> moving faster towards the solution
- But Batch GD is faster in terms of latency

#### Loss Plot
- Loss is always smoooth in Batch GD
- But Fluctuating/Noisy in Stochastic GD (bcoz loss reduction is unstable in Stochastic)

1. Stochastic GD

```python
model.compile(loss='binary_crossentropy',metrics=['accuracy'])
# start = time.time()
history = model.fit(X_scaled,y,epochs=500,batch_size=1,validation_split=0.2)
# print(time.time() - start)
```

```python
import matplotlib.pyplot as plt
plt.plot(history.history['loss'])
```

2. Batch GD

```python
model.compile(loss='binary_crossentropy',metrics=['accuracy'])
# start = time.time()
history = model.fit(X_scaled,y,epochs=500,batch_size=len(X_scaled),validation_split=0.2)
# print(time.time() - start)
```

```python
import matplotlib.pyplot as plt
plt.plot(history.history['loss'])
```
